```python
import scipy
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.misc
import scipy.signal
import matplotlib.image as mpimg
```

```python
image = cv2.imread('bb.jpg',0)
image.shape
#print(image.shape)
#print(image.shape)
#cv2.imshow('sample image',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#sharpen image 
sharp = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharpimg = scipy.signal.convolve2d(image,sharp,mode='same')
sharpimg = (sharpimg+255)/2

plt.imshow(image,cmap='gray')
plt.show()
plt.imshow(sharpimg,cmap='gray')
plt.show()
```

```python
#sobel operator
sobel = dict()
sobel["x"] = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel["y"] = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
G_x = scipy.signal.convolve2d(image,sobel["x"])
G_y = scipy.signal.convolve2d(image,sobel["y"])
print(image.shape)
print(G_x.shape)
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# Actually plt.imshow() can handle the value scale well even if I don't do 
# the transformation (G_x + 255) / 2.
ax1.imshow((G_x + 255) / 2, cmap='gray'); ax1.set_xlabel("Gx")
ax2.imshow((G_y + 255) / 2, cmap='gray'); ax2.set_xlabel("Gy")
plt.show()
```
